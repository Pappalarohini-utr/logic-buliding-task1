```python
# Problem 1: Employee Performance Bonus Eligibility
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Find the highest performance score
max_score = max(employees.values())

# Find all employees with the highest score
top_performers = [name for name, score in employees.items() if score == max_score]

# Display the result
print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {max_score})\n")

```

```python
# Problem 2: Search Query Keyword Analysis
search_query = "Buy mobile phone buy phone online"

# Convert to lowercase and split into words
words = search_query.lower().split()

# Count word frequencies
word_freq = {}
for word in words:
    # Remove punctuation if needed (simple version)
    clean_word = ''.join(char for char in word if char.isalnum())
    word_freq[clean_word] = word_freq.get(clean_word, 0) + 1

# Filter words with frequency > 1
repeated_keywords = {word: count for word, count in word_freq.items() if count > 1}

print(repeated_keywords, "\n")
```

```python

sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# Collect valid readings (even numbers)
valid_readings = [(index, value) for index, value in enumerate(sensor_readings) if value % 2 == 0]

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings, "\n")
```

```python
# Problem 4: Email Domain Usage Analysis
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Count domain usage
domain_count = {}
for email in emails:
    domain = email.split('@')[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

# Calculate percentage
total_emails = len(emails)
for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {int(percentage)}%")
print()

```

```python
# Problem 5: Sales Spike Detection
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average sales
avg_sales = sum(sales) / len(sales)

# Detect spikes (>30% above average)
for day, sale in enumerate(sales, start=1):
    if sale > 1.3 * avg_sales:
        print(f"Day {day}: {sale}")
print()
```

```python
# Problem 6: Duplicate User ID Detection
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Count occurrences
id_count = {}
for uid in user_ids:
    id_count[uid] = id_count.get(uid, 0) + 1

# Filter duplicates and display
for uid, count in id_count.items():
    if count > 1:
        print(f"{uid} → {count} times")
```
